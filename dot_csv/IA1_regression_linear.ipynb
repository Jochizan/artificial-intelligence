```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
# Cargamos las librerias
boston = load_boston()

print(boston.DESCR)
```

Formula para minimizar el error cuadrático medio (MCO): $\beta = (X^{T}X)^{-1}X^{T}Y$

```python
train_data = np.array(boston.data[:, 5])
target_data = np.array(boston.target)

# Datos de entrenamiento
# print(train_data)

# Datos de salida
# print(target_data)

# Gráfica de nuestras variables

plt.scatter(train_data, target_data, alpha=0.4)

# Añadimos columna de 1s para el termino independiente
train_data = np.array([np.ones(506), train_data]).T

beta = np.linalg.inv(train_data.T @ train_data) @ train_data.T @ target_data

# print(transpose)

plt.plot([4, 9], [beta[0] + beta[1] * 4, beta[0] + beta[1] * 9], c='red')
plt.show()
```
